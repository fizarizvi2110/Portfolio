### Writing your own k-means clustering algorithm

K-means clustering is a simple unsupervised machine-learning method for cluster analysis. The aim of the method is to partition a set of points into k clusters, such that each point is assigned to the nearest cluster. The algorithm iterates through two steps:

1. Assign each data point to the cluster with the nearest centroid
2. Update the centroids of the clusters given the new assignment

The algorithm converges when the assignments no longer change. Since the intial assignment to clusters is largely random, there is no guarantee that the optimum assignment is found. So it is common to run the algorithm multiple times and use different starting conditions.

In this problem set, we will implement a much simplified version of the k-means clustering algorithm. Rather than running the algorithm until convergence, we will repeat the above two steps a large but fixed number of times. In addition, we will initialize only once, using a naive method according to which we randomly choose k points from the data to use as initial cluster centroids.

### Problem 1: Function to estimate Euclidean distance between two points


```python
def get_distance(x, y):
    """Calculates euclidean distance between two n-dimensional points 
    stored in lists and returns its value in a float
    """
    
    # Var to track sum of squared differences
    sum_squared_diff = 0 

    if(len(x)) != (len(y)):
        print("Error: Both points must have equal number of dimensions.")

    # Summing up squared differences for each dimension in both lists
    else:
        for i in range(len(x)):
            sum_squared_diff += (x[i] - y[i]) ** 2
        return math.sqrt(sum_squared_diff)


```

### Problem 2: Function to estimate the centroid of a collection of points

```python
def get_centroid(points):
    """Estimates and returns the centroid of a list of n-dimensional points contained
    in a list
    """

    num_points = len(points)

    # Number of dimensions of point, starting at first point
    num_dimens = len(points[0])

    # Calculation of mean of each dimension in each point stored in list
    return [sum(point[i] for point in points) / num_points for i in range(num_dimens)]


```

### Problem 3: Function to read data

```python


def get_data(f_loc):
    """Reads in csv file using file path and returns list of the data for each row 
    contained in a list, skipping non numerical data
    """

    # Open and read csv file, skipping the header row
    with open(f_loc, newline='') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)

        # Skip first two columns of each row and convert rest to float value
        return [[float(val) for val in row[2:]] for row in reader]


```

### Problem 4: Function to implement k-means algorithm

function called `kmeans` that clusters a collection of points into k clusters using a simplified version of the k-means algorithm. The function has two parameters: 

1. `points` – a list of n-dimensional points;
2. `k` – an integer that defines the number of desired clusters.

```python

def kmeans(points, k):
    """Clusters a list of n-dimensional points into k number of clusters using a k-means
    algorithm implementation with a set 100 iterations, and returns a tuple containing clusters and 
    their centroids
    """
    
    # Select k random points to use as initial centroids
    init = random.sample(points, k)

    # Create a list of k lists to contain the points assigned to each cluster  
    clusters = [[] for i in init]
    
    # Create a list to keep the centroids of the k clusters
    # For now, this list will contain the points from init
    centroids = [i for i in init]
    
    
    for i in range(100):
        
        # Clear previous clusters to reassign points to new updated centroids
        clusters = [[] for i in range(k)]
    
        for point in points: 
            
            # Calculating distance from each point to all centroids in a list
            distances = [get_distance(point, centroid) for centroid in centroids]

            # Finding closest centroid by getting smallest distance of point from each centroid
            closest_centroid = distances.index(min(distances))

            # Assigning the point to the cluster of the closest centroid
            clusters[closest_centroid].append(point)

        # Calculates new updated centroid for each updated cluster 
        centroids = [get_centroid(cluster) for cluster in clusters]

    return clusters, centroids
```
